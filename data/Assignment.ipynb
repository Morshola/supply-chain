```python
# import pandas
import pandas as pd
```

```python
# Read csv
warehouse_reporting = pd.read_csv("warehousing-reporting.csv")
warehouse_reporting.head()
```

```python
# Check csv information
warehouse_reporting.info()
```

```python
# Convert "DATE OF LAST ORDER" to a datetime type. (maybe lookup astype(datetime))
warehouse_reporting["date_of_last_order"] = warehouse_reporting["date_of_last_order"].astype("datetime64[ns]")
warehouse_reporting.info()
```

```python
# Convert "TOTAL VALUE" to a float.
warehouse_reporting["total_value"] = warehouse_reporting["total_value"].str.replace("$","")
warehouse_reporting["total_value"] = warehouse_reporting["total_value"].str.replace(",","")
warehouse_reporting["total_value"] = warehouse_reporting["total_value"].astype("float")
warehouse_reporting.info()
```

```python
# Add a column which calculates the "Reorder Status" from the "STOCK QUANTITY" and "REORDER LEVEL" columns. 
warehouse_reporting["reorder_status_new"] = warehouse_reporting["stock_quantity"] < warehouse_reporting["reorder_level"]
warehouse_reporting.head()
```

```python
warehouse_reporting[["reorder_(auto-fill)", "reorder_status_new"]].value_counts()
```

```python
# Split the "STOCK LOCATION" column so "Store Room A" and "Shelf 2" are in different columns. 
warehouse_reporting[["stock_location_1", "stock_location_2"]] = warehouse_reporting["stock_location"].str.split(',', expand=True)
warehouse_reporting.head()
```
